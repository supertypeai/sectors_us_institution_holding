```python
import datetime
filing_data = nc.select_query("SELECT filing_id, cik, filing_date, accession_number FROM form_13f_filing")
print(len(filing_data))

for filing in filing_data:
    df = find(filing['accession_number']).obj().infotable.sort_values('Value', ascending = False)[['Ticker','Value','SharesPrnAmount']]
    df['filing_id'] = filing['filing_id']
    df['filing_date'] = filing['filing_date'] 
    df['filing_date'] = df['filing_date'].apply(lambda x: x.strftime('%Y-%m-%d %H:%M:%S'))
    df[['Value','SharesPrnAmount']] = df[['Value','SharesPrnAmount']].astype(int)
    df = df.groupby(['Ticker']).sum().reset_index().sort_values('Value', ascending = False)
    total_shares = df['Value'].sum()
    df['percentage'] = df['Value'] / total_shares * 100
    df = df.rename(columns={'Ticker': 'symbol', 'Value': 'value','SharesPrnAmount':'share'})
    json_string = df.to_json(orient='records')
    recs = json.loads(json_string)
    nc.batch_upsert(target_table="form_13f_holdings", records=recs, conflict_columns=['id'])
```

```python
filing_data = nc.select_query("SELECT filing_id, cik, filing_date, accession_number FROM form_13f_filing")
newest_filings = {}

# Iterate through the data and update the newest filing for each CIK
for filing in filing_data:
    cik = filing['cik']
    filing_date = filing['filing_date']
    if cik not in newest_filings or filing_date > newest_filings[cik]['filing_date']:
        newest_filings[cik] = filing

# Extract the newest filings as a list
newest_filings_list = list(newest_filings.values())
newest_filings_list
```
